```python
import pandas as pd

def merge_by_uid(rainfall_csv_path, ndvi_csv_path, output_csv_path):
    """
    Create a new CSV with UID, rainfall_mm, and ndvi_value by left joining on UID.
    
    Args:
        rainfall_csv_path: Path to the rainfall CSV file (contains all UIDs)
        ndvi_csv_path: Path to the NDVI CSV file
        output_csv_path: Path to save the merged output CSV
    """
    
    # Read both CSV files
    rainfall_df = pd.read_csv(rainfall_csv_path)
    ndvi_df = pd.read_csv(ndvi_csv_path)
    
    # Perform left join to merge NDVI values into rainfall data
    merged_df = rainfall_df[['UID', 'rainfall_mm']].merge(
        ndvi_df[['UID', 'ndvi_value']],
        on='UID',
        how='left'
    )
    
    # Save the merged dataframe to a new CSV file
    merged_df.to_csv(output_csv_path, index=False)
    print(f"Merged CSV saved to {output_csv_path}")
    print(f"Total rows: {len(merged_df)}")
    print(f"Rows with NDVI values: {merged_df['ndvi_value'].notna().sum()}")

# Example usage:
merge_by_uid(
    rainfall_csv_path="Datasets/GC_final/final_rainfall.csv",
    ndvi_csv_path="Datasets/GC_final/final_ndvi.csv",
    output_csv_path="rainfall_ndvi_merged.csv"
)
```

```python
import pandas as pd

def merge_datasets(rainfall_csv_path, ndvi_csv_path, lst_csv_path, output_csv_path):
    """
    Create a new CSV with UID, rainfall_mm, ndvi_value, and lst_value by left joining on UID.
    
    Args:
        rainfall_csv_path: Path to the rainfall CSV file (contains all UIDs)
        ndvi_csv_path: Path to the NDVI CSV file
        lst_csv_path: Path to the LST CSV file
        output_csv_path: Path to save the merged output CSV
    """
    
    # Read all CSV files
    rainfall_df = pd.read_csv(rainfall_csv_path)
    ndvi_df = pd.read_csv(ndvi_csv_path)
    lst_df = pd.read_csv(lst_csv_path)
    
    # Perform the first left join: rainfall + NDVI
    merged_df = rainfall_df[['UID', 'rainfall_mm']].merge(
        ndvi_df[['UID', 'ndvi_value']],
        on='UID',
        how='left'
    )
    
    # Perform the second left join: merged_df + LST
    merged_df = merged_df.merge(
        lst_df[['UID', 'lst_value']],
        on='UID',
        how='left'
    )
    
    # Save the final merged dataframe to a new CSV file
    merged_df.to_csv(output_csv_path, index=False)
    print(f"Merged CSV saved to {output_csv_path}")
    print(f"Total rows: {len(merged_df)}")
    print(f"Rows with NDVI values: {merged_df['ndvi_value'].notna().sum()}")
    print(f"Rows with LST values: {merged_df['lst_value'].notna().sum()}")

# Example usage:
merge_datasets(
    rainfall_csv_path="Datasets/GC_final/final_rainfall.csv",
    ndvi_csv_path="Datasets/GC_final/final_ndvi.csv",
    lst_csv_path='Datasets/GC_final/final_lst.csv',
    output_csv_path='rainfall_ndvi_lst_merged.csv'
)
```

```python
import pandas as pd

def remove_incomplete_rows(input_csv_path, output_csv_path):
    """
    Create a new CSV file by removing rows with any missing values.
    
    Args:
        input_csv_path: Path to the input CSV file (merged dataset with potential missing values)
        output_csv_path: Path to save the cleaned output CSV file
    """
    
    # Read the input CSV file
    df = pd.read_csv(input_csv_path)
    
    # Drop rows with any missing values
    cleaned_df = df.dropna()
    
    # Save the cleaned dataframe to a new CSV file
    cleaned_df.to_csv(output_csv_path, index=False)
    print(f"Cleaned CSV saved to {output_csv_path}")
    print(f"Total rows after cleaning: {len(cleaned_df)}")

# Example usage:
remove_incomplete_rows(
    input_csv_path='rainfall_ndvi_lst_merged.csv',
    output_csv_path='rainfall_ndvi_lst_cleaned.csv'
)
```
